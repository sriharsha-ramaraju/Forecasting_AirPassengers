##### Let us use another visualisation technique, rolling mean and std to visualise how mean is changing over time

```python
ts_rollmean=ts.rolling(window=12).mean()
ts_rollstd=ts.rolling(window=12).std()

plt.plot(ts,label="Original")
plt.xlabel('Years')
plt.ylabel('Number of Passengers')
plt.plot(ts_rollmean,label='Rolling Mean')
plt.plot(ts_rollstd,label='Rolling Std')
plt.legend(loc="best")
plt.show()
```

##### The above visualisation shows that, series is not stationary, however, let's prove it using Augumented Dicker-Fullery Test as one can see, mean is increasing over the time and standard deviation is also increasing

##### Agumented Dicker-Fullery test

###### The Augmented Dickey-Fuller test is a type of statistical test called a unit root test.

The intuition behind a unit root test is that it determines how strongly a time series is defined by a trend.

There are a number of unit root tests and the Augmented Dickey-Fuller may be one of the more widely used. It uses an autoregressive model and optimizes an information criterion across multiple different lag values.

The null hypothesis of the test is that the time series can be represented by a unit root, that it is not stationary (has some time-dependent structure). The alternate hypothesis (rejecting the null hypothesis) is that the time series is stationary.

Null Hypothesis (H0): If failed to be rejected, it suggests the time series has a unit root, meaning it is non-stationary. It has some time dependent structure.
Alternate Hypothesis (H1): The null hypothesis is rejected; it suggests the time series does not have a unit root, meaning it is stationary. It does not have time-dependent structure.

```python
from statsmodels.tsa.stattools import adfuller
def is_stationary(time_series):
    x=adfuller(time_series)
    print("Test-Statistic:",round(x[0],3))
    print("p-value:",round(x[1],3))
    if x[1]<0.05:
        print("Time Series is STATIONARY")
    else:
        print("Time Series is NOT STATIONARY")
        
```

```python
is_stationary(ts['#Passengers'])
```

##### Lets make the data to be stationary..there are many ways we will be using, 1)log transform 2)substraction of moving avg 3)difference operator

```python
ts_log=np.log(ts)#method 1

mvAvg=ts_log.rolling(window=12).mean()
ts_logMinusmvAvg=ts_log-mvAvg #method 2

ts_logdiff=ts_log-ts_log.shift() #method 3

```

#### LOG TRANSFORM

```python
#plt.plot(ts,label="original")
plt.plot(ts_log,label="log-transform")
plt.plot(ts_log.rolling(window=12).mean(),label="Rolling Mean")
plt.plot(ts_log.rolling(window=12).std(),label="Rolling StdDev")
#plt.plot(ts_logMinusmvAvg,label="log-mvAvg")
#plt.plot(ts_logdiff,label="log-diff")
plt.xlabel('Years')
plt.ylabel('Number of Passengers')
plt.legend(loc='best')
plt.show()

is_stationary(ts_log['#Passengers'])
```

##### taking "log" of time series is not turning it into stationary, however, it's far better than the original ts

#### SUBSTRACTION OF MOVING AVERAGE

```python
#plt.plot(ts,label="original")
plt.plot(ts_logMinusmvAvg,label="log-mvAvg")
plt.plot(ts_logMinusmvAvg.rolling(window=12).mean(),label="Rolling Mean")
plt.plot(ts_logMinusmvAvg.rolling(window=12).std(),label="Rolling StdDev")
plt.xlabel('Years')
plt.ylabel('Number of Passengers')
plt.legend(loc='best')
plt.show()


#is_stationary(ts_logMinusmvAvg['#Passengers'])#running this code gives you error as ts contains nans..
#which needs to be dropped

print("Before Removing NaNs")
print(ts_logMinusmvAvg.head())

print("After Removing NaNs")
ts_logMinusmvAvg.dropna(inplace=True)
print(ts_logMinusmvAvg.head())

print("----------------------")

is_stationary(ts_logMinusmvAvg['#Passengers'])#checking for stationarity
```

##### In the plot above, rolling mean and rolling std seems to be parallel to x-axis, ie constant with time. The stationarity us also supported by DK test with a p-value of 0.02

#### DIFFERENCE OPERATOR

```python
#plt.plot(ts,label="original")
plt.plot(ts_logdiff,label="")
plt.plot(ts_logdiff.rolling(window=12).mean(),label="Rolling Mean")
plt.plot(ts_logdiff.rolling(window=12).std(),label="Rolling StdDev")
plt.xlabel('Years')
plt.ylabel('Number of Passengers')
plt.legend(loc='best')
plt.show()


#is_stationary(ts_logMinusmvAvg['#Passengers'])#running this code gives you error as ts contains nans..
#which needs to be dropped

print("Before Removing NaNs")
print(ts_logdiff.head())

print("After Removing NaNs")
ts_logdiff.dropna(inplace=True)
print(ts_logdiff.head())

print("----------------------")

is_stationary(ts_logdiff['#Passengers'])#checking for stationarity
```

#### Although DK-test shows the TS to be non-stationary, the p-value is much closer to 0.05 and visually the mean is stationary

### Let us use ARIMA to forecast the series..to use use ARIMA the time series has to be stationary

```python
from pmdarima.arima import auto_arima
model=auto_arima(ts)
warnings.filterwarnings('ignore')
```

```python
model
```

##### So the best ARIMA model is of order 2,1,2 i.e p(auto regressive-pacf)=2,d(difference)=1,q(moving average-acf)=2

#### So let's check acf and pacf plots to make sure we are using correct order

```python
from statsmodels.tsa.seasonal import seasonal_decompose

from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.graphics.tsaplots import plot_pacf


plot_acf(ts)
plot_pacf(ts)
plt.show()
```

```python
from statsmodels.tsa.arima_model import ARIMA
model_arima=ARIMA(ts,order=(2,1,2))
model_arima_fit=model_arima.fit()

plt.plot(ts.diff(),label="Original")
plt.plot(model_arima_fit.fittedvalues,label="Fitted values",color="r")
plt.legend(loc="best") #method 3,label="Fitted values")
plt.show()
```

```python
print(model_arima_fit.fittedvalues.shape)
print(ts[1:].diff().shape)
```

```python


ts_diff=ts.diff(periods=1)#differencing
print(ts_diff.head())#visialising ts after differencing

ts_diff.dropna(inplace=True)#dropping nan
print("---------------")
print("After removing NaN")
print(ts_diff.head())#visialising ts after removing nan

print("---------------")
print("Fitted Values")
print(model_arima_fit.fittedvalues.head())

```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error

def calc_metrics(actual,predicted):
    mse=mean_squared_error(actual,predicted)
    mae=mean_absolute_error(actual,predicted)
    print("MSE:", round(mse,2))
    print("MAE:", round(mae,2))
    print("RMSE:",round(np.sqrt(mse),2))
```

```python
calc_metrics(ts_diff,model_arima_fit.fittedvalues)
```

#### Let us use HOLT WINTERS model

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing

model_hw=ExponentialSmoothing(ts,seasonal_periods=12,seasonal="mul",trend="add").fit()
```

```python
plt.plot(ts,label="Original")
plt.plot(model_hw.fittedvalues,label="fittedvalues")
plt.legend(loc="best")
plt.show()

calc_metrics(ts,model_hw.fittedvalues)
```

### From the above two methods, its clear that Holt-Winters performs really well in comparision to ARIMA, however, we used entire data to build the model and test on the same data which is not the right way to do, so let us split the data into training and testing, lets leave last 6months for test data

```python
train_ts,test_ts=ts[:120],ts[120:] #first 138 values for training and last 138 values for test
model_train_arima=auto_arima(train_ts,trace=True).fit(train_ts)#throws best order (3,1,3), using this order let's build arima model
forecast=model_train_arima.predict(len(test_ts)) #forecasting for next 6 periods (6months)
```

```python
plt.plot(test_ts,label="test")
plt.plot(train_ts,label="train")
plt.plot(test_ts.index,forecast,label="forecast",color="red")
plt.plot(train_ts.index,model_train_arima.predict_in_sample(),label="fittedvalues")
plt.legend(loc="best")
plt.show()

print("Train Error Metrics:")
calc_metrics(train_ts,model_train_arima.predict_in_sample())
print("----------------------------------")

print("Forecast Error Metrics:")
calc_metrics(test_ts,forecast)
```

#### From the above figure, it seems that ARIMA seems to be poorly fitting the test data, although it has failry performed on the train data let us apply HOLT WINTERS AND CHECK IIF IT PERFORMS WELL


```python
model_train_hw=ExponentialSmoothing(train_ts,seasonal_periods=12,seasonal="mul",trend="add").fit()
forecast_hw=model_train_hw.predict(start=test_ts.index[0],end=test_ts.index[-1])
```

```python
plt.plot(test_ts,label="test")
plt.plot(train_ts,label="train")
plt.plot(forecast_hw,label="forecast",color="red")
plt.plot(model_train_hw.fittedvalues,label="fittedvalues")
plt.legend(loc="best")
plt.show()

print("Train Error Metrics:")
calc_metrics(train_ts,model_train_hw.fittedvalues)
print("----------------------------------")

print("Forecast Error Metrics:")
calc_metrics(test_ts,forecast_hw.values)
```
